```python
def getIndex(dfObj, value):
    ''' Get the index position of value in the dataframe dfObj.'''
 
    listOfPos = list()
    # Get bool dataframe with True at positions where the given value exists
    result = dfObj.isin([value])
    # Get list of columns that contains the value
    seriesObj = result.any()
    columnNames = list(seriesObj[seriesObj == True].index)
    # Iterate over list of columns and fetch the rows indexes where value exists
    for col in columnNames:
        rows = list(result[col][result[col] == True].index)
        for row in rows:
            listOfPos.append((row, col))
    # Return a list of tuples indicating the positions of value in the dataframe
    return listOfPos
```

```python
def getEleve():
    eleves = pd.read_excel(entree_eleves_apres_pretraitement)
    eleves = eleves.fillna("")
    eleves = eleves.rename(columns={'Destinataires': 'Nom'})
    eleves['HORAIRE 1'] = 0
    eleves['HORAIRE 2'] = 0
    eleves['HORAIRE 3'] = 0
    eleves['PARENT 1'] = 0
    eleves['PARENT 2'] = 0
    eleves['PARENT 3'] = 0
    eleves['CONVOCATION'] = 0
    return eleves
```

```python
def getParent():
    parents = pd.read_excel(entree_parents_apres_pretraitement)
    parents = parents.rename(columns={'Participant':'Nom'})
    parents[1] = 0; parents[2] = 0; parents[3] = 0; parents[4] = 0; parents[5] = 0; parents[6] = 0;
    return parents
```

```python
def getBestSlot(parent, index, slotInUse1, slotInUse2, slotInUse3):
    '''return the slot with minimum students except if already in use'''
    nbCrenaux = parent.loc[index,'creneaux']
    debut = parent.loc[index,'debut']
    ibest = 1000
    minimumNumberOfStudents = 1000
    # for i in range(1,nbCrenaux+1):
    #     if i != slotInUse1 and i != slotInUse2 and i != slotInUse3:
    #         minimumNumberOfStudents = parent.loc[index,i]
    #         ibest = i
    #         break
    for i in range(debut,nbCrenaux+debut):
        if i != slotInUse1 and i != slotInUse2 and i != slotInUse3:
            numberOfStudents = parent.loc[index,i]
            if numberOfStudents < minimumNumberOfStudents:
                minimumNumberOfStudents = numberOfStudents
                ibest = i
    return ibest
```

```python
def getLessBusyParent(parents, listOfPos):
    """Retourne le parent le moins occupé
    L'entrée est le dataframe des parents et une liste de tuple (index, col)
    
    Index étant l'index d'un parent avec le métier choisi"""

    nombre_entretiens = []
    for index_parent,_ in listOfPos:
        nombre_entretiens.append( (parents.loc[index_parent, range(1,7)].sum()) / (parents.loc[index_parent, "creneaux"]))
    index_minimum = nombre_entretiens.index(min(nombre_entretiens))

    return listOfPos[index_minimum][0]
```

```python
creneaux = {1: '9h35->9h55',
            2: '10h05->10h25',
            3: '10h35->10h55',
            4: '11h05->11h25',
            5: '11h35->11h55'}
```

```python
'''Main loop'''
for index, row in eleves.iterrows():
    entretiens = ['Entretien 1', 'Entretien 2', 'Entretien 3']
    horaires = ['HORAIRE 1', 'HORAIRE 2', 'HORAIRE 3']
    parents_horaires = ['PARENT 1', 'PARENT 2', 'PARENT 3']
    for i in range(3):
        metier=row[entretiens[i]]
        
        # Donne une liste de tuples (index, col) des parents ayant le métier metier
        listOfPos = getIndex(parents[['Secteur']],metier)
        index_parent = getLessBusyParent(parents, listOfPos)
        # print(listOfPos, index_parent)
        bestSlot = getBestSlot(parents, index_parent, eleves.loc[index,horaires[0]], 
                               eleves.loc[index,horaires[1]], eleves.loc[index,horaires[2]])
        # print(index_parent, listOfPos, bestSlot, eleves.loc[index, horaires])
        parents.loc[index_parent,bestSlot] += 1
        eleves.at[index,horaires[i]] = bestSlot
        eleves.at[index,parents_horaires[i]] = index_parent
    eleves.at[index,'CONVOCATION'] = min(eleves.loc[index,horaires[0]], 
                                         eleves.loc[index,horaires[1]], 
                                         eleves.loc[index,horaires[2]])
```

```python
def ecrire_entretien(index, row):
    if index == 0:
        return f"{creneaux[row['HORAIRE 1']]}: {row['Entretien 1']} ({parents.loc[row['PARENT 1'], 'Nom']}, {parents.loc[row['PARENT 1'], 'Métier']})\n"
    elif index == 1:
        return f"{creneaux[row['HORAIRE 2']]}: {row['Entretien 2']} ({parents.loc[row['PARENT 2'], 'Nom']}, {parents.loc[row['PARENT 2'], 'Métier']})\n"
    elif index == 2:
        return f"{creneaux[row['HORAIRE 3']]}: {row['Entretien 3']} ({parents.loc[row['PARENT 3'], 'Nom']}, {parents.loc[row['PARENT 3'], 'Métier']})\n"
    else:
        raise ValueError("Index must be between 0 and 2; however, it is {index}")
        
```

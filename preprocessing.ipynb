Il fait le preprocessing du fichier 'entree_eleves', qui contient une colonne par métier/secteur.

```python
import pandas as pd
import numpy as np
```

```python
entree_eleves_path = "../entree/Répartition def eleves  participant V2.xlsx"
entree_parents_path = "../entree/Fichier participants forum des métiers 2024.xlsx"
```

```python
eleves_src = pd.read_excel(entree_eleves_path)
# eleves_src = eleves_src.drop(labels=[0])
eleves_src.head()
```

```python
# We first filter out unwanted rows and columns

eleves_src = pd.read_excel(entree_eleves_path)
eleves_src = eleves_src.drop(labels=[0, 1, 2, 3], axis="index")
eleves_src = eleves_src.drop(columns=["répondu"])
eleves_src = eleves_src.drop(eleves_src.filter(regex='Unnamed:').columns, axis="columns")
eleves_src = eleves_src.dropna(how="all")
list_classes = [301, 302, 303, 304, 305]
eleves_src = eleves_src[eleves_src["Classe"].isin(list_classes)]
eleves_src[-10:]
```

```python
# Rearrange columns abd filters NaN

eleves = eleves_src.copy()
eleves["Destinataires"] = eleves["Nom"] + " " + eleves["Prénom"]
eleves = eleves.drop(columns=["Nom", "Prénom"])
eleves = eleves.set_index('Destinataires')
eleves = eleves.fillna(0)
eleves = eleves.replace('x',1)
eleves = eleves.replace('X',1)
eleves[-3:]
```

```python
eleves_out = pd.DataFrame(index=eleves.index)
eleves_out["Classe"] = ""
eleves_out["Entretien 1"] = ""
eleves_out["Entretien 2"] = ""
eleves_out["Entretien 3"] = ""
eleves_out["Entretien 4"] = ""
eleves_out.head()
```

```python
for i in range(len(eleves)):
    a = eleves[i:(i+1)]
    # print(a)
    b = a.loc[:, (a>0).any()]
    c = b.columns.tolist()
    if len(c) < 5:
        c.append("")
    if len(c) < 4:
        print(a.index, len(c))
    eleves_out[i:(i+1)] = c
eleves_out['Classe'] = eleves['Classe'] 
eleves_out.head()
```

```python
eleves_out.to_excel("../entree/entree_eleves_après_prétraitement.xlsx")  
```

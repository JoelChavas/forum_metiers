Il fait le preprocessing du fichier 'entree_eleves', qui contient une colonne par métier/secteur.

```python
import pandas as pd
import numpy as np
```

```python
eleves_src = pd.read_excel('entree_eleves.xlsx')
eleves_src = eleves_src.drop(labels=[0])
eleves_src.head()
```

```python

eleves = eleves_src[:147].drop(['répondu', 'Designer', 'Nombre de choix', 'Les métiers de l\'artisanat', 'Les métiers du cinéma (Couturière styliste)', 'Les métiers de l\'artisanat / Jardinier'], axis=1)

eleves = eleves.rename(columns={'Classes':'Classe'})
eleves.head()
```

```python
eleves[-3:]
```

```python
eleves.columns
```

```python
len(eleves.columns)
```

```python
eleves = eleves.fillna(0)
eleves = eleves.replace('x',1)
eleves = eleves.replace('X',1)
```

```python
eleves = eleves.set_index('Destinataires')
```

```python
eleves.head()
```

```python
eleves_out = pd.DataFrame(index=eleves.index)
eleves_out["Classe"] = ""
eleves_out["Entretien 1"] = ""
eleves_out["Entretien 2"] = ""
eleves_out["Entretien 3"] = ""
eleves_out["Entretien 4"] = ""
eleves_out.head()
```

```python
for i in range(len(eleves)):
    a = eleves[i:(i+1)]
    # print(a)
    b = a.loc[:, (a>0).any()]
    c = b.columns.tolist()
    if len(c) < 5:
        c.append("")
    if len(c) < 4:
        print(a.index, len(c))
    eleves_out[i:(i+1)] = c
eleves_out['Classe'] = eleves['Classe'] 
eleves_out.head()
```

```python
eleves_out.to_excel("entretien_eleves.xlsx")  
```
